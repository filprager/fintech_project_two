```python
import requests
import os
import json

# To set your environment variables in your terminal run the following line:
# export 'BEARER_TOKEN'='<your_bearer_token>'


def auth():
    #return os.environ.get("BEARER_TOKEN")
    return os.environ.get("TWITTER_BEARER_TOKEN")


def create_url(userid):
    # Replace with user ID below
    user_id = userid
    return "https://api.twitter.com/2/users/{}/tweets".format(user_id)


def get_params():
    # Tweet fields are adjustable.
    # Options include:
    # attachments, author_id, context_annotations,
    # conversation_id, created_at, entities, geo, id,
    # in_reply_to_user_id, lang, non_public_metrics, organic_metrics,
    # possibly_sensitive, promoted_metrics, public_metrics, referenced_tweets,
    # source, text, and withheld
    # return {"tweet.fields": "created_at"}
    return {"tweet.fields": "created_at", "max_results":100 }


def create_headers(bearer_token):
    headers = {"Authorization": "Bearer {}".format(bearer_token)}
    return headers


def connect_to_endpoint(url, headers, params):
    response = requests.request("GET", url, headers=headers, params=params)
    print(response.status_code)
    if response.status_code != 200:
        raise Exception(
            "Request returned an error: {} {}".format(
                response.status_code, response.text
            )
        )
    return response.json()


def main():
    userid1=44196397 #elon musk
    bearer_token = auth()
    url = create_url(userid1)
    print(url)
    headers = create_headers(bearer_token)
    params = get_params()
    json_response = connect_to_endpoint(url, headers, params)
    #print(json.dumps(json_response, indent=4, sort_keys=True))
    #print(json_response)
    twit_dict = json_response['data']
    
    #print(twit_dict)
    twit_df = pd.DataFrame(twit_dict)
    display(twit_df)
    return twit_df

if __name__ == "__main__":
    main()
```

```python
import requests
import os
import json

# To set your environment variables in your terminal run the following line:
# export 'BEARER_TOKEN'='<your_bearer_token>'


def auth():
    #return os.environ.get("BEARER_TOKEN")
    return os.environ.get("TWITTER_BEARER_TOKEN")


def create_url(userid):
    # Replace with user ID below
    user_id = userid
    return "https://api.twitter.com/2/users/{}/tweets".format(user_id)


def get_params(next_token):
    # Tweet fields are adjustable.
    # Options include:
    # attachments, author_id, context_annotations,
    # conversation_id, created_at, entities, geo, id,
    # in_reply_to_user_id, lang, non_public_metrics, organic_metrics,
    # possibly_sensitive, promoted_metrics, public_metrics, referenced_tweets,
    # source, text, and withheld
    # return {"tweet.fields": "created_at"}
    return {"max_results": 100,
        "tweet.fields": "created_at",
        "expansions": "author_id" , 'pagination_token': next_token}


def create_headers(bearer_token):
    headers = {"Authorization": "Bearer {}".format(bearer_token)}
    return headers


def connect_to_endpoint(url, headers, params):
    response = requests.request("GET", url, headers=headers, params=params)
    print(response.status_code)
    if response.status_code != 200:
        raise Exception(
            "Request returned an error: {} {}".format(
                response.status_code, response.text
            )
        )
    return response.json()





def main():
    userid1=44196397 #elon musk
    
    next_token= None
    bearer_token = auth()
    url = create_url(userid1)
    
    #print(url)
    headers = create_headers(bearer_token)
    #print('headers\n',headers)
    params = get_params(next_token)
    #print('params\n',params)
    json_response = connect_to_endpoint(url, headers, params)
    #print(json.dumps(json_response, indent=4, sort_keys=True))
    #print(json_response)
    twit_dict = json_response['data']
    next_token = json_response['meta']['next_token']
    #print('-----------')
    #print('next token: ' , next_token)
    twit_df = pd.DataFrame(twit_dict)
    display(twit_df)
    #return twit_df

if __name__ == "__main__":
    main()

'''
This function iterates across pages of tweets and gets all tweets for a single
user. Pass in user id.
'''
def getUserTweets(userid):
    
    # base dataframe
    twit_df = pd.DataFrame()
    
    
    # set iteration continuation to true initially
    hasNextPage = True
    
    # set initial nexttoken
    nextToken = None

    # while there are tokens continue to extract
    tmpa = 0
    while hasNextPage:
        
        # get the url 
        bearer_token = auth()
        url = create_url(userid1)
        
        # minimum sleep between iterations is 5 seconds
        sleep(7)
        
        #print(url)
        headers = create_headers(bearer_token)
        #print('headers\n',headers)
        params = get_params(nextToken)
        #print('params\n',params)
        json_response = connect_to_endpoint(url, headers, params)
        #print(json.dumps(json_response, indent=4, sort_keys=True))
        #print(json_response)
        twit_dict = json_response['data']

        twit_df_tmp = pd.DataFrame(twit_dict)
        twit_df = pd.concat([twit_df,twit_df_tmp])
        if debug:
            print('length of dataframe: ', len(twit_df))
        display(twit_df)
        
        try:
            nextToken = json_response['meta']['next_token']
            print('-----------')
            print('nextToken token: ' , nextToken)
        except:
            # no more tokens
            hasNextPage = False
            
        tmpa +=1 # increase backup counter
        
        # set a maximum as a backup
        if tmpa==9000:
            hasNextPage = False
            
    if debug:
        display(twit_df)
        print(len(twit_df))        
    return  twit_df
```
